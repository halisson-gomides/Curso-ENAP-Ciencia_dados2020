```python
# Passo 5: Junção de informações dos Beneficiários com as Tabelas Auxiliares do Cadastro de Pessoas Físicas da RFB.
# É criada a Tabela WD_AUX_EMERG que será transformada em DataFrame Pandas. O Join com o Cadastro de Pessoas Físicas
# é feito pelos 6 Dígitos centrais do CPF que está mascarado no Portal da Transparência e pelo Nome do Beneficiário.
sql = '''
CREATE TABLE IF NOT EXISTS u14422131168.WD_AUX_EMERG STORED AS PARQUET AS
SELECT
    em.ano_mes,
	em.uf AS uf_orig,
	pf.nr_unidade_federacao_pf AS uf_nr,
	uf.SG_UNIDADE_FEDERACAO AS uf_sg,
    em.munic_orig,
	pf.nr_municipio_pf AS munic_nr,
	mun.NM_MUNICIPIO AS munic_nm,
    SUBSTR(REPLACE(em.cpf_orig,".",""),4,6) AS cpf_orig,
	pf.nr_cpf AS cpf_cad,
    em.nome_orig,
	pf.nm_contrib_pf,
	SUBSTR(REPLACE(em.cpf_resp,".",""),4,6) AS cpf_resp,
    em.enquadramento,
    STRLEFT(em.parcela,1) AS parcela,
    em.vlr_benef,
	pf.cd_reg_polit_pf,
	rp.nm_regiao_politica,
	pf.nr_sit_cadastral_pf,
	sc.nm_sit_cadastral,
	pf.nr_ocup_prin_pf,
	op.nm_ocup_prin,
	op.nr_classe_ocup_prin,
	op.nm_classe_ocup_prin,
	op.nr_grupo_ocup_prin,
	op.nm_grupo_ocup_prin,
	op.nr_agrega_ocup_prin,
	op.nm_agrega_ocup_prin,
	pf.ano_nasc_pf,
	pf.ano_obito_pf,
	pf.nr_pf_sexo,
	sx.nm_pf_sexo,
	pf.nr_pf_ind_estrangeiro,
	ie.nm_pf_ind_estrangeiro,
	pf.nr_pf_porte,
	po.nm_pf_porte,
	pf.nr_sit_cadin,
	ca.nm_sit_cadin
FROM
	u14422131168.202007_auxilioemerg AS em
	JOIN DIME.WD_PESSOAS_FISICAS AS pf
	ON SUBSTR(REPLACE(em.cpf_orig,".",""),4,6) = SUBSTR(pf.nr_cpf,4,6)	AND em.nome_orig = pf.nm_contrib_pf
	JOIN DIME.WD_LC_UNIDADES_FEDERACAO AS uf 
	ON pf.nr_unidade_federacao_pf = uf.NR_UNIDADE_FEDERACAO 
	JOIN DIME.WD_LC_MUNICIPIOS AS mun 
	ON	pf.nr_municipio_pf = mun.NR_MUNICIPIO
	JOIN u14422131168.WD_LC_REGIOES_POLITICAS AS rp
	ON pf.cd_reg_polit_pf = rp.cd_regiao_politica
	JOIN u14422131168.WD_PF_SIT_CADASTRAL AS sc
	ON pf.nr_sit_cadastral_pf = sc.nr_sit_cadastral
	JOIN u14422131168.WD_OP_OCUPACOES_PRINCIPAIS AS op
	ON pf.nr_ocup_prin_pf = op.nr_ocup_prin
	JOIN u14422131168.WD_PF_SEXO AS sx
	ON pf.nr_pf_sexo = sx.nr_pf_sexo
	JOIN u14422131168.WD_PF_IND_ESTRANGEIRO AS ie
	ON pf.nr_pf_ind_estrangeiro = ie.nr_pf_ind_estrangeiro
	JOIN u14422131168.WD_PF_PORTE AS po
	ON pf.nr_pf_porte = po.nr_pf_porte
	JOIN u14422131168.WD_SIT_CADIN AS ca
	ON pf.nr_sit_cadin = ca.nr_sit_cadin
;
COMPUTE STATS u14422131168.WD_AUX_EMERG   -- cria índices e referências para otimizar as consultas
'''
pd.read_sql(sql, cnx)
```

```python
sql = ''' 
SELECT 
    COUNT(*) AS qte, 
    SUM(vlr_benef) AS valor
FROM u14422131168.WD_AUX_EMERG;'''
df_ref = pd.read_sql(sql, cnx)
qte = df_ref['qte'][0]
vlr = df_ref['valor'][0]
print('qte_linhas = ', qte, ' (17,59 Milhões)        soma_benefícios = R$', vlr.astype(float), ' (R$ 11,6 Bilhões)')

# Obs: a tabela WD_AUX_EMERG resultante tem quantidade menor de linhas pq não foi encontrada correspondência do cpf6 e nome
# em cerca de 900 mil CPFs Beneficiários com as tabelas do Cadastro PF da RFB. O estudo seguirá com este número reduzido.
```

```python
# Passo 8: Seleção de colunas do DataFrame para estudo de correlação (redução para viabilizar a consulta de 17,5 Mi linhas)
# A partir dos 16 agrupamentos de informações acima, selecionou-se:
#      - nm_regiao_politica;
#      - nm_grupo_ocup_prin;
#      - idade;
#      - faixa_idade;
#      - nm_pf_ind_estrangeiro;
#      - nm_pf_sexo;
#      - UFs de amostragem: 'DF', 'GO'
#  Obs: foi necessário reduzir substancialmente o dataset por razões de desempenho e tráfego de rede

# SQL do Auxílio Emergencial (ae):
sql_ae = ''' 
SELECT
	em.cpf_orig AS cpf,
	em.vlr_benef,
	pf.nr_unidade_federacao_pf AS uf_nr,
	uf.SG_UNIDADE_FEDERACAO AS uf,
	rp.cd_regiao_politica,
	rp.nm_regiao_politica AS regiao,
	op.nr_agrega_ocup_prin,
	op.nm_agrega_ocup_prin,
	(2020 - pf.ano_nasc_pf) AS idade,
	CASE 
        WHEN (2020 - pf.ano_nasc_pf) >= 60 THEN 60
        WHEN (2020 - pf.ano_nasc_pf) >= 50 THEN 50
        WHEN (2020 - pf.ano_nasc_pf) >= 40 THEN 40
        WHEN (2020 - pf.ano_nasc_pf) >= 30 THEN 30
        WHEN (2020 - pf.ano_nasc_pf) >= 20 THEN 20
        ELSE 10
    END AS faixa_idade,
	pf.nr_pf_ind_estrangeiro,
	ie.nm_pf_ind_estrangeiro AS estrangeiro,
	pf.nr_pf_sexo,
	sx.nm_pf_sexo AS sexo
FROM
	u14422131168.202007_auxilioemerg AS em
	JOIN DIME.WD_PESSOAS_FISICAS AS pf
	ON SUBSTR(REPLACE(em.cpf_orig,".",""),4,6) = SUBSTR(pf.nr_cpf,4,6)	AND em.nome_orig = pf.nm_contrib_pf
	JOIN DIME.WD_LC_UNIDADES_FEDERACAO AS uf 
	ON pf.nr_unidade_federacao_pf = uf.NR_UNIDADE_FEDERACAO 
	JOIN u14422131168.WD_LC_REGIOES_POLITICAS AS rp
	ON pf.cd_reg_polit_pf = rp.cd_regiao_politica
	JOIN u14422131168.WD_OP_OCUPACOES_PRINCIPAIS AS op
	ON pf.nr_ocup_prin_pf = op.nr_ocup_prin
	JOIN u14422131168.WD_PF_SEXO AS sx
	ON pf.nr_pf_sexo = sx.nr_pf_sexo
	JOIN u14422131168.WD_PF_IND_ESTRANGEIRO AS ie
	ON pf.nr_pf_ind_estrangeiro = ie.nr_pf_ind_estrangeiro
WHERE 
    uf.SG_UNIDADE_FEDERACAO IN ('DF', 'RS', 'PE', 'AM')
'''
```

```python
# DataFrame do Auxílio Emergencial (ae)
df_ae = pd.read_sql(sql_ae, cnx)
df_ae.head()
```
